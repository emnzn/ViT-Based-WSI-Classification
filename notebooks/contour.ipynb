```python
import os

import cv2
import pyvips
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = os.path.join("..", "data", "images")
image_names = os.listdir(data_dir)

print("num images:", len(image_names))
```

```python
image_paths = [os.path.join(data_dir, i) for i in image_names]

images = [pyvips.Image.new_from_file(f, access="sequential") for f in image_paths]
```

```python
test_image = images[1].numpy()

# width, height = test_image.shape[1], test_image.shape[0]

# test_image = cv2.resize(test_image, (width // 10, height // 10))
```

```python
gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.show()
```

```python
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
x_min, y_min, x_max, y_max = float('inf'), float('inf'), 0, 0

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    if w > 500 and h > 500:
        x_min = min(x_min, x)
        y_min = min(y_min, y)
        x_max = max(x_max, x + w)
        y_max = max(y_max, y + h)
```

```python
contoured_image = cv2.rectangle(test_image, (x_min, y_min), (x_max, y_max), (255, 0, 0), 200)

plt.imshow(contoured_image)
plt.axis("off")
plt.show()
```
