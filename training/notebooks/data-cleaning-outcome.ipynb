```python
import os
import re
from pathlib import Path

import numpy as np
import pandas as pd
```

```python
data_dir = os.path.join("..", "..", "raw-data", "clinical", "outcome.xlsx")

os.path.isfile(data_dir)
```

```python
pd.ExcelFile(data_dir).sheet_names
```

```python
df = pd.read_excel(data_dir, sheet_name='Sheet1')
df["S/N"] = df["S/N"].fillna(method="ffill")
df["id"] = df["S/N"].map(int).map(str)

df = df.drop(["S/N", "NRIC"], axis=1)
df = df[["id", "recur", "timetorecur"]]

df.head()
```

```python
img_dir = os.path.join("..", "..", "raw-data", "images")

ids = [i.split(".")[0] for i in os.listdir(img_dir)]
ids = list(set([re.sub("\D", "", i) for i in ids]))
ids[:5]
```

```python
len(ids)
```

```python
df = df[df["id"].isin(ids)]

df
```

```python
df[df["recur"].isna()]
```

```python
missing_ids = {
    "id": [i for i in ids if i not in df["id"].tolist()],
    "recur": [np.nan for _ in range(8)],
    "timetorecur": [np.nan for _ in range(8)]
    }
```

```python
df1 = pd.DataFrame(missing_ids)
```

```python
pd.concat([df[df["recur"].isna()], df1]).reset_index(drop=True)
```
